```python
X = data.loc[:, data.columns != "MedHouseVal"]
y = data['MedHouseVal']
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X = scalar.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(activation='relu',hidden_layer_sizes=(130,64,32,16))
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,r2_score
r2_score(y_test, predictions)
```

```python
model.score(X_train, y_train)
```

```python
model = MLPRegressor(activation='relu',hidden_layer_sizes=(13,13,13,13))
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
r2_score(y_test, predictions)
```

```python
# My model didnt perform that well hopefully with GridSearch would provide better data
```

```python
model.score(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param = {
    'activation':["identity", "logistic", "tanh", "relu"] ,
    'solver':["lbfgs", "sgd", "adam"],
    "learning_rate":["constant", "invscaling", "adaptive"]
}
clf = GridSearchCV(MLPRegressor(hidden_layer_sizes=(13,13,13,13)),param)

clf.fit(X_train, y_train)
print(clf.score(X_train, y_train))
print(clf.best_params_)
```

```python
# Well my accuracy went up and the best parameters is
# 'activation': 'tanh', 'learning_rate': 'constant', 'solver': 'adam'
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
iris_df.loc[:, iris_df.columns != "species"] = scaler.fit_transform(iris_df.loc[:, iris_df.columns != "species"])
train, test = train_test_split(iris_df, test_size = 0.3)
trainX = train[['sepal_len','sepal_wid','petal_len','petal_wid']]# taking the training data features
trainY=train.species# output of our training data
testX= test[['sepal_len','sepal_wid','petal_len','petal_wid']] # taking test data features
testY =test.species   #output value of test data
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(activation ='relu',hidden_layer_sizes=(256,128,64,32))
```

```python
clf.fit(trainX, trainY)
```

```python
predictions = clf.predict(testX)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(testY,predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(testY,predictions))
print(classification_report(testY,predictions))
cm = confusion_matrix(testY,predictions)
```

```python
clf = MLPClassifier(activation ='relu',hidden_layer_sizes=(130,64,32,16))
clf.fit(trainX, trainY)
predictions = clf.predict(testX)
accuracy_score(testY,predictions)
```

```python
clf = MLPClassifier(activation ='relu',hidden_layer_sizes=(13,13,13,13))
clf.fit(trainX, trainY)
predictions = clf.predict(testX)
accuracy_score(testY,predictions)
```

```python
clf = GridSearchCV(MLPClassifier(hidden_layer_sizes=(13,13,13,13)),param)

clf.fit(trainX, trainY)
print(clf.score(trainX, trainY))
print(clf.best_params_)
```
