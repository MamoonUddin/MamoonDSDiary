```python

# Name two different models that you can use to solve the problem statement.
# What is the difference between label encoding and one-hot encoding, 
# and when should you use one over the other?
# What is multicollinearity, and why do we care about it when creating models? 
# How can we check to see if theres significant multicollinearity in our data?
# Why is scaling data important?
# For the two different models that you named earlier, are they using a parametric or 
# non-parametric learning algorithm? Whats the difference?
# Suppose that we had missing values in our dataset. What are different ways we could 
# handle them

# I decided to use Linear Regression and Random Forest Classifier. \
# Label Encoding, each label is assigned a unique integer based on alphabetical ordering.
# Meanwhile, One-Hot Encoding simply creates additional features based on the number of unique values in the categorical feature
# And we use the label when there is order in the data and we use one-hot encoding when there is no order.
# We can check multi-coplanarity by making a heat map or making scatterplots and checking it manually.
# Feature scaling is essential for machine learning algorithms that calculate distances between data. 
# If not scale, the feature with a higher value range starts dominating when calculating distances
# The logistic regression model is parametric and the random forest is a non-parametric model.
# Parametric statistics are based on assumptions about the distribution of the population from which the sample was taken.
# Nonparametric statistics are not based on assumptions
# To handle our missing data we would do amputations or drop the columns as a whole. Amputations are when you
# replace the missing data with respectable values such as the mean.
```

```python
from sklearn.model_selection import train_test_split

X = data_df.loc[:, data_df.columns != "output"]
y = data_df['output']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
rfc_pred = rfc.predict(X_test)
print('Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, rfc_pred)))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rfc, X_train, y_train, cv = 10, scoring='accuracy')
print('Cross validation scores:{}'.format(scores))

print('\nAverage cross validation score: {:.4f}'.format(scores.mean()))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, rfc_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, rfc_pred)
print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0,0])
print('\nTrue Negatives(TN) = ', cm[1,1])
print('\nFalse Positives(FP) = ', cm[0,1])
print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
# now for logistic regression

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = data_df.drop('output', axis=1)
y = data_df['output']

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X = pd.DataFrame(data=X_scaled, columns=X.columns)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
cm = confusion_matrix(y_test, predictions)
print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0,0])
print('\nTrue Negatives(TN) = ', cm[1,1])
print('\nFalse Positives(FP) = ', cm[0,1])
print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(logmodel, X_train, y_train, cv = 10, scoring='accuracy')
print('Cross validation scores:{}'.format(scores))

print('\nAverage cross validation score: {:.4f}'.format(scores.mean()))
```
